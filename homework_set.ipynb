# Bài tập 2.1.15


Use a Venn diagram to illustrate the set of all months of the year whose names do not contain the letter R in theset of all months of the year.

```python
months = {'January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December'}
month_not_r = set()
for single_month in months:
    if 'R' not in single_month.upper():
        month_not_r.add(single_month)
print(month_not_r)
```

```python
venn2([set(months), 
               set(month_not_r)], ('Month: '+ ', '.join(months), 'Month dont have R letter: '+ ", ".join(month_not_r))) # 3-element list of set objects

plt.show()
```

# Bài 2.1.24

Can you conclude that $A = B$ if $A$ and $B$ are two sets with the same power set?

Explain: 
$A = B$ => $P(A) = P(B)$. Prove it?

```python
A = {'Binh', 'Duc', 'Loi'}
B = {'Loi', 'Duc', 'Binh'}
```

```python


def powerset(iterable):
    s = list(iterable)
    return list(chain.from_iterable(combinations(s, r) for r in range(len(s)+1)))
```

```python
powerset_A = powerset(A)
print(powerset_A)
powerset_B = powerset(B)
print(powerset_B)

if (powerset_A == powerset_B):
    print("Powerset bằng nhau")
else: 
    print("Powerset không bằng nhau")
```
